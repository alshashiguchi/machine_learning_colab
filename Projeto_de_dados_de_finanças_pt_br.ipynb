```
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
```

```
start = datetime.datetime(2008,1,1)
end = datetime.datetime(2019,1,1)
PETR4 = data.DataReader("PETR4.SA", "yahoo", start, end)
USIM5 = data.DataReader("USIM5.SA", "yahoo", start, end)
ABEV3 = data.DataReader("ABEV3.SA", "yahoo", start, end)
ITUB4 = data.DataReader("ITUB4.SA", "yahoo", start, end)
BBDC4 = data.DataReader("BBDC4.SA", "yahoo", start, end)
VALE3 = data.DataReader("VALE3.SA", "yahoo", start, end)
```

```
BBDC4.head()
```

```
BBDC4['Close'].plot()
```

```
tickers = ['PETR4', 'USIM5', 'ABEV3', 'ITUB4', 'BBDC4', 'VALE3']
```

```
tickers
```

```
bank_stocks = pd.concat([PETR4, USIM5, ABEV3, ITUB4, BBDC4, VALE3], axis=1, keys=tickers)
```

```
bank_stocks.head()
```

```
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```
bank_stocks.head()
```

```
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

```
returns = pd.DataFrame()
```

```
returns.head()
```

```
for tick in tickers:
  returns[tick + 'Return']=bank_stocks[tick]['Close'].pct_change()
```

```
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```
for tick in tickers:
  bank_stocks[tick]['Close'].plot(figsize=(12,4), label=tick)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```
bank_stocks.xs(key='Close', axis=1, level='Stock Info').plot()
```

```
PETR4Return2016 = PETR4['Close'][PETR4.index.year==2016]
PETR4Return2016_media_movel = PETR4Return2016.rolling(window=30).mean()
PETR4Return2016_media_movel.head(50)
```

```
plt.figure(figsize=(12,6))
PETR4Return2016_media_movel.plot(label='Média móvel de 30 dias')
PETR4Return2016.plot(label='Fechamento')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
